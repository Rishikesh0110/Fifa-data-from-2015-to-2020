###Analyzing column Age

```
print(fifa_19["age"].mean()) #Statistical analysis of age column
print(fifa_19["age"].median())
print(fifa_19["age"].max())
print(fifa_19["age"].min())
print(fifa_19["age"].std())
print(fifa_19["age"].var())
print(fifa_19["age"].skew())
```

```
plt.style.use('ggplot') ##Histogram
plt.figure(figsize=(10,8))
plt.title("Histogram for age",size=20)
sns.histplot(x=fifa_19["age"],bins=20,color="red")
plt.xlabel("age",size=12)
plt.ylabel("frequency",size=12)
plt.show()
```

```
plt.style.use('ggplot')
plt.figure(figsize=(12,8))
#plt.title("distribution for age",size=20)
sns.displot(x=fifa_19["age"],kde="kind",color="red")
plt.xlabel("age",size=12)
plt.ylabel("frequency",size=12)
plt.show()
```

```
plt.style.use('ggplot') #Box plot
plt.figure(figsize=(12,8))
plt.title("boxplot for age",size=20)
sns.boxplot(x=fifa_19["age"],color="gold",orient="h")
plt.xlabel("age",size=12)
plt.ylabel("frequency",size=12)
plt.show()
```

```
plt.style.use('ggplot') #violin plot
plt.figure(figsize=(10,6))
plt.title("violin for age",size=20)
sns.violinplot(x=fifa_19["age"],color="red")
plt.xlabel("age",size=12)
plt.ylabel("frequency",size=12)
plt.show()
```

```
plt.style.use('ggplot')
plt.figure(figsize=(12,6))
ax = sns.violinplot(data=fifa_19, x='age', color="#af52f4", inner=None, linewidth=0, saturation=0.9)
sns.boxplot(data=fifa_19, x='age', saturation=0.5, width=0.4,
            palette='rocket', boxprops={'zorder': 2}, ax=ax)
plt.show()
```

Conclusion
1. So we can see that data is centred around mean value 25.193922341024198 the median value 25.0
2. The maximum age value is 45 the min value is 16 for fifa 19.
3. We  can see that standard daviation value 4.65 with variance  21.640 for age column.
4. Also age column is symmetrical in nature.
